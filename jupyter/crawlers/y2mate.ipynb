```python
tbody = table.find('tbody')
```

```python
trs = tbody.find_all('tr')
len(trs)
```

```python
import re 

def select_file(resolution='720p'):
    for tr in tbody.find_all('tr'):
        for td in tr.children:
            m = re.search(f'({resolution})', string=td.text)
            if m is not None: 
                return td.parent

tr = select_file()
tr 
```

```python
tr.find_all('td')[2].button.attrs
```

```python
onclick_value = tr.find_all('td')[2].button.attrs['onclick']
onclick_value
```

```python
def extract_download_id(tr):
    # 3번째 td가 다운로드 버튼이다.
    onclick_value = tr.find_all('td')[2].button.attrs['onclick']
    m = re.search("'(mp4)',\s*'(.+)'", string=onclick_value)
    print(m)
    print(m.groups())
    return m.group(2)

dl_id = extract_download_id(tr)
dl_id
```

```python
dl_button = driver.find_element(By.XPATH, f"//button[contains(@onclick, '{dl_id}')]")
dl_button
```
